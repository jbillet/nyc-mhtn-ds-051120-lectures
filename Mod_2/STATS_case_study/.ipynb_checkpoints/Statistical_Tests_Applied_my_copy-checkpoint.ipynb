## 1. Handle the non-normal values

As you can see from the data dictionary, there are a number of responses that are coded with values that could throw off our descriptive statistics.  Let's clean up by mapping the responses to a more appropriate value, and handling any missing values.

```python
# creating dictionary with the values we want to replace in ALCDAY5
alc_dict = {888:0, 
            777: np.nan,
            999: np.nan
            }
# creating new column
df['alcohol'] = df['ALCDAY5'].replace(alc_dict)
```

```python
# creating function to continue to clean values 
def calc_monthly(x):
    if x>200:
        return x-200
    elif x>100:
        return min((x-100)*4.34524, 30)
    else:
        return x
df['alcohol']= df['alcohol'].apply(calc_monthly)
```

- Create a 95% confidence interval for the population mean of physically ill days.

```python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m-h, m+h
```

```python
mean_confidence_interval(df['PHYSHLTH'],confidence=0.95)
```
